## <font color=red> IMPORTING THE LIBRARIES </font>

```python
import datetime as dt
df["date"]=pd.to_datetime(df["date"])
df["date"]=df["date"].map(dt.datetime.toordinal)
```

```python
df['tests_units'] = df['tests_units'].replace([np.nan,'samples tested'],[0,1]).astype('float')
```

## <font color=red>CLEANING THE DATA AND DROPPING OR REPLACING THE NULL VALUES </font>

```python
df.isnull().sum()
```

### <font color=red>HERE WE OBSERVE THAT WE HAVE NULL VALUES IN COLUMNS </font>

#### - total_tests----85
#### - new_tests------91
#### - total_tests_per_thousand ---85
#### - new_tests_per_thousand------91
#### - new_tests_smoothed----------80
#### - new_tests_smoothed_per_thousand----80
#### - stringency_index------3

### <font color=red>- THEREFORE IT IS CLEARLY OBSERVED THAT THIS IS NOT MORE THAN 50% AS WE HAVE SEEN FROM THE SHAPE HENCE WELL BE REPLACING IT WITH THE MEAN, MEDIAN AND MODE. </font>

```python
df['total_tests']=df['total_tests'].fillna(df['total_tests'].median())

df['new_tests']=df['new_tests'].fillna(df['new_tests'].mean())

df['total_tests_per_thousand']=df['total_tests_per_thousand'].fillna(df['total_tests_per_thousand'].median())

df['new_tests_per_thousand']=df['new_tests_per_thousand'].fillna(df['new_tests_per_thousand'].mean())

df['new_tests_smoothed']=df['new_tests_smoothed'].fillna(df['new_tests_smoothed'].mean())

df['new_tests_smoothed_per_thousand']=df['new_tests_smoothed_per_thousand'].fillna(df['new_tests_smoothed_per_thousand'].mean())

df['stringency_index']=df['stringency_index'].fillna(df['stringency_index'].median())
```

```python
df.isna().sum()
```

## <font color=red>SELECTING THE TARGET COLUMN AS TOTAL_CASES AND REST OF THE COLUMNS AS FEATURES</font>

```python
X = df.drop('total_cases',axis=1)
y = df['total_cases']
```

## <font color=red>DIVIDING THE DATA INTO TRAIN AND TEST DOING TRAIN TEST SPLIT</font>

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## <font color=red>LINEAR REGRESSION</font>

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
sns.distplot((y_test-predictions),bins=30);
```

```python
predictions = lm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('Variance score: %.2f' % lm.score(X_test, y_test))
```

```python
lm.score(X_test,y_test)
```

## <font color=red>RANDOM FOREST</font>
